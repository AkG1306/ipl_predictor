### batter_stats

```python
batters = balls.groupby(["batter"])
batter_stats = pd.DataFrame(batters['ball'].count()).rename(columns={"ball":"balls_faced"})
batter_stats
```

```python
batter_stats['innings'] = batters['inning'].nunique()
batter_stats['runs'] = batters['batsman_runs'].sum()
batter_stats
```

```python
batter_stats["0s"] = balls[balls["batsman_runs"]==0].groupby("batter")["batsman_runs"].count()
batter_stats['0s'].fillna(0,inplace=True)

batter_stats["1s"] = balls[balls["batsman_runs"]==1].groupby("batter")["batsman_runs"].count()
batter_stats['1s'].fillna(0,inplace=True)

batter_stats["2s"] = balls[balls["batsman_runs"]==2].groupby("batter")["batsman_runs"].count()
batter_stats['2s'].fillna(0,inplace=True)

batter_stats["3s"] = balls[balls["batsman_runs"]==3].groupby("batter")["batsman_runs"].count()
batter_stats['3s'].fillna(0,inplace=True)

batter_stats["4s"] = balls[balls["batsman_runs"]==4].groupby("batter")["batsman_runs"].count()
batter_stats['4s'].fillna(0,inplace=True)

batter_stats["6s"] = balls[balls["batsman_runs"]==6].groupby("batter")["batsman_runs"].count()
batter_stats['6s'].fillna(0,inplace=True)
```

```python
batter_stats
```

```python
batter_stats["player_out"] = batters["player_dismissed"].count()
```

```python
batter_stats["strike_rate"] = round(batter_stats["runs"]/batter_stats["balls_faced"],2)
batter_stats["strike_rate"].fillna(0,inplace = True)
batter_stats["batting_avg"] = round(batter_stats["runs"]/batter_stats["player_out"],2)
batter_stats["batting_avg"].fillna(0,inplace = True)
```

```python
batter_stats
```

### bowler_stats

```python
bowlers = balls.groupby("bowler")
bowler_stats = pd.DataFrame(bowlers['ball'].count()).rename(columns={'ball':'balls_thrown'})
```

```python
wickets = balls[balls["dismissal_kind"].isin(['caught','bowled', 'lbw','stumped', 'caught and bowled', 'hit wicket'])]
bowler_stats["wickets"] = wickets.groupby(["bowler"])["batter"].count()
bowler_stats["wickets"].fillna(0,inplace= True)
```

```python
bowler_stats["overs"] = round(bowler_stats["balls_thrown"]/6,2)
```

```python
bowler_stats["runs_conceded"] = bowlers["total_runs"].sum()
bowler_stats["runs_conceded"].fillna(0,inplace = True)
```

```python
bowler_stats["economy"] = round(bowler_stats["runs_conceded"]/bowler_stats["overs"],2)
bowler_stats["economy"].fillna(0,inplace = True)
```

```python
bowler_stats["bowling_strike_rate"] = round(bowler_stats["balls_thrown"]/bowler_stats["wickets"],2)
bowler_stats["bowling_strike_rate"].fillna(0,inplace = True)
bowler_stats["bowling_strike_rate"] = np.where(bowler_stats["wickets"]==0,0,bowler_stats["bowling_strike_rate"])
```

## Players and no_of_matches


```python
players_matches_dict = {}

# Iterate over each row in the dataframe
for i, row in balls.iterrows():
    # Check if the batter is already in the dictionary
    if row['batter'] in players_matches_dict:
        players_matches_dict[row['batter']].add(row['match_id'])
    else:
        players_matches_dict[row['batter']] = {row['match_id']}

    # Check if the non-striker is already in the dictionary
    if row['non_striker'] in players_matches_dict:
        players_matches_dict[row['non_striker']].add(row['match_id'])
    else:
        players_matches_dict[row['non_striker']] = {row['match_id']}

    # Check if the bowler is already in the dictionary
    if row['bowler'] in players_matches_dict:
        players_matches_dict[row['bowler']].add(row['match_id'])
    else:
        players_matches_dict[row['bowler']] = {row['match_id']}

# Create a dataframe with players and their number of matches
final_players_matches = pd.DataFrame({'Players': list(players_matches_dict.keys())})
final_players_matches['matches'] = final_players_matches['Players'].apply(lambda x: len(players_matches_dict[x]))
final_players_matches = final_players_matches.set_index('Players')
```

```python
caught_and_bowled = balls[balls["dismissal_kind"]=="caught and bowled"].groupby("bowler")["player_dismissed"].count()
caught_feilder = balls[balls["dismissal_kind"]=="caught"].groupby("fielder")["player_dismissed"].count()

catch_df = pd.merge(caught_and_bowled,caught_feilder, left_index=True, right_index=True,how='outer')
catch_df.fillna(0,inplace=True)
catch_df['catches'] = catch_df['player_dismissed_x']+catch_df['player_dismissed_y']
catch_df.drop(['player_dismissed_x','player_dismissed_y'],axis=1,inplace=True)
```
